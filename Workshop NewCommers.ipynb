# Go to ACALIB/acalib/cupid and run 'make'

```python
import acalib
```

```python
c = acalib.Container()
```

```python
c.load_fits('bindata/fits/cubes/ALMA01000740.fits')
```

```python
#Vemos la cantidad de imagenes
len(c.images)
```

```python
#Y la cantidad de tablas
len(c.tables)
```

```python
#Obtenemos el cubo de datos
cube = c.primary
```

```python
#Momento 0 (Collapsando el eje de las frecuencias)
mom0 = acalib.moment0(cube)
plt.imshow(mom0.data)
```

```python
#Estimating rms (root mean square) and thresholding
rms = acalib.rms(cube)
cleaned = acalib.denoise(cube, threshold=rms)
```

```python
#Momento 0 (Collapsando el eje de las frecuencias)
mom0_c = acalib.moment0(cleaned)
plt.imshow(mom0_c.data)
```

```python
#Otra forma de graficar una imagen 2D
acalib.io.visualize_image(mom0)
acalib.io.visualize_image(mom0_c)
```

## PyCupid

```python
fw = acalib.algorithms.FellWalker()
caa, clumps = fw.run(cube)
```

```python
dims = caa.data.shape
for i in range(dims[0]):
    if caa.data[i,:,:].sum() > 0:
        plt.imshow(caa.data[i,:,:])
        plt.show()
```
